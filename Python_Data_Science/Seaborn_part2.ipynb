```python
### Importing the seaborn libraray
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.shape
```

### Categorical Plot --- 
1. VilonPlot 
2. count Plot
3. Bar Plot


```python
sns.countplot(x='day',data=df)
```

```python
sns.countplot(y='sex',data=df)  # Thisis only used for categorical analysis
```

### BarPlot
here we have to give both the axis but in countplot we don't have to give

```python
sns.barplot(x='total_bill',y='smoker',data=df)
```

```python
sns.barplot(x='sex',y='tip',data=df)   ## One categorical and one numerica; l
```

### Box Plot
here we don't give the specific axis. we can print graph by giving aal the data

```python
# sns.boxplot(x='sex',y='total_bill',data=df)
sns.boxplot(data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

### Violin Polt 
it is the combination of the kde and box plot 

```python
sns.violinplot(x='total_bill',y='day',data=df)
```
