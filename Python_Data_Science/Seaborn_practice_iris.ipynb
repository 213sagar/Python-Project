```python
### ImPorting the library of seaborn and numpy
import seaborn as sns
import numpy as np
```

```python
dataset=sns.load_dataset('iris')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.info()
```

### Jointplot


```python
sns.jointplot(x='sepal_length',y='sepal_width',data=dataset,kind='hex')
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=dataset,kind='reg')
```

###   BOX Plot

```python
sns.boxplot(data=dataset)
```

```python
sns.boxplot(x='species',y='sepal_width',data=dataset)
```

### CountPlot

```python
## It will give the count of eac categorical variable --- Here we have to give the only categorical variable
sns.countplot(x='species',data=dataset,palette='rainbow')
```

## Distplot - it is similar to histogram 


```python
sns.displot(data=dataset,kde=True)
```

### Violin Plot 

```python
sns.violinplot(x='petal_width',y='species',data=dataset,vert=False)
```

```python
### Bar Plot
sns.barplot(x='species',y='sepal_length',data=dataset)
```
