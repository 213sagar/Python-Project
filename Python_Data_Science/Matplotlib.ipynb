### Matplotlib
Matplotlib is a cross-platform, data visualization and graphical plotting library for Python and its numerical extension NumPy. As such, it offers a viable open source alternative to MATLAB.

```python
### Importing Matplotlib library
import matplotlib.pyplot as plt
%matplotlib inline  
## This line is used to avoil plt.show() line
```

```python
#Import numpy libraray
import numpy as np
```

```python
##Simple examples 

a=np.arange(0,10)
v=np.arange(10,20)
ca=np.arange(23,33)
d=np.arange(33,43)
```

```python
### Ploting using matplotlib 
### Scatter Plot
plt.scatter(a,v,c='g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot')
```

```python
### Plot function 
y=a*a
plt.plot(a,y,'r*--')
```

```python
### Creating Subplot
plt.subplot(2,2,1)
plt.plot(a,v,'r')
plt.subplot(2,2,2)
plt.plot(y,v,'g*')
plt.subplot(2,2,3)
plt.plot(a,d,'b-*')
plt.subplot(2,2,4)
plt.plot(ca,d,'r*')
```

```python
### Generation of sin and cos wave
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave generation')
plt.plot(x,y)
# print(type(x))
```

```python
cos_wave=np.cos(x)
plt.title('Cos wave generation')
plt.plot(x,cos_wave,'r*-')
```

### Barplot

```python
plt.bar(a,a,color='g')
plt.title('Bar Plot Creation')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

### Histograms

```python
###hIstograms --- 
a=np.array([23,24,48,52,59,32,32,14,55,34,130,3])
plt.hist(a,bins=10)
plt.title('histogram Chart creation')
```

### bOX Plot

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
##Rectangular  box plot
plt.boxplot(data,vert=True,patch_artist=True)

```
