### What is Seaborn Library
Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.

This is a updated version of the matplotlib we are often use this to visualize tha data. We use this to anlyze the multiple features. 

### Different Types of plots are there - 

####  This plots is applicable on Discrete Values
1. displot 
2. JoinPlot  --- it is used to do the anlayze  bivariate
3. Pairplot ---  if we have to work multiple dimention then we have to use this 

```python
### import seaborn library
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
##total_bill		sex	smoker	day	time	size === Independent Features
## tip - DEpendent features
```

```python
df.dtypes
```

```python
# Correlation with hEatmap--- 
df.corr()
```

```python
sns.heatmap(df.corr())   ### It will give visualiztion correlation status with different color
```

## Join Plot

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### Pairplot

```python
sns.pairplot(df,hue='smoker') ### If we have to visualize more than two feature and find the  relation we use PairPlot
```

## Dist PLOt 

```python
sns.displot(df['tip'],color='g')  ## It is similar to  histograms in matplotlib
```

```python
sns.displot(df['tip'],kde='True') ### iT is plot with  kernal density function
```
